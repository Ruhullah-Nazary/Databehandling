# Laboration_Databehandling
## Del A

```python
import pandas as pd

# Läs csv filen med namnet "College"
college = pd.read_csv("college.csv") 
# Visa de fem första raderna
college.head()
```

---
## Del B

```python
# Byt namn på kolumnen
college.rename(columns={"Unnamed: 0": "College"}, inplace=True)
# Sätt "College" som index
college.set_index("College", inplace=True)
college.head()
```

---
## Del C

```python
# Generera en sammanfattning av statistiken av numeriska data.
summary = college.describe() 
summary
```

---
## Del D

```python
import matplotlib.pyplot as plt

# Välj dessa kolumner
columns_to_plot = ["Top10perc", "Apps", "Enroll"] 
# Plotta och skapa scatter matrix.
pd.plotting.scatter_matrix(college[columns_to_plot], alpha=0.8, figsize=(10, 10), diagonal="hist", c="green") 
plt.show()
```

---     
## Del E

```python
# Skapa en boxplot av "Outstate" mot "Private"
college.boxplot (column="Outstate", by="Private") 

# Skapa title och etiketter 
plt.title("Outstate Tuition by Private College Status")
plt.suptitle(" ") 
plt.xlabel("Private College")
plt.ylabel("Outstate Tuition")
plt.show()
```

---
## Del F

```python
# Skapa "Elite" som är baserad på kolumnen "Top10perc"
college["Elite"] = pd.cut(college["Top10perc"], bins=[0, 50, 100], labels=["No", "Yes"], right=True)

# Räkna hur många "Elite" universitet finns det
elite_counts = college["Elite"].value_counts()
print(elite_counts)

# Skapa boxplott av "Outstate" mot "Elite"
college.boxplot(column="Outstate", by="Elite")

# Skapa title och etiketter 
plt.title('Outstate Tuition by Elite College Status')
plt.suptitle('')
plt.xlabel('Elite Status')
plt.ylabel('Outstate Tuition')
plt.show()
```

---
## Del G

```python
# Skapa en subplot
fig, axes = plt.subplots(2, 2, figsize=(13, 10))

# Skapa histogram för fyra kvantitativa variabler med olika lagerstorlekar
college["Outstate"].plot.hist(bins=15, ax=axes[0, 0], color="lightblue", edgecolor="black")
axes[0, 0].set_title("Outstate Tuition (15 bins)")
axes[0, 0].set_xlabel("Outstate Tuition")

college["Apps"].plot.hist(bins=20, ax=axes[0, 1], color="lightgreen", edgecolor="black")
axes[0, 1].set_title("Applications (20 bins)")
axes[0, 1].set_xlabel('Applications')

college["Enroll"].plot.hist(bins=25, ax=axes[1, 0], color="lightcoral", edgecolor="black")
axes[1, 0].set_title("Enrollment (25 bins)")
axes[1, 0].set_xlabel("Enrollment")

college["Room.Board"].plot.hist(bins=30, ax=axes[1, 1], color="lightyellow", edgecolor="black")
axes[1, 1].set_title("Room and Board (30 bins)")
axes[1, 1].set_xlabel("Room and Board")

plt.show()
```

---
## Del H

```python
import seaborn as sns
# Korrelationsanalys
correlation_matrix = college[["Personal", "Expend", "Outstate", "Top10perc"]].corr()
print(correlation_matrix)

# Rita upp korrelationsmatrisen
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar=True)
plt.title("Correlation Matrix")
plt.show()

# Scatter Plots for "Personal" mot "Expend" och "Personal" mot "Elite"
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Personal", y="Expend", data=college, hue="Elite", palette="coolwarm")
plt.title("Personal Spending vs Instructional Spending")
plt.xlabel('Personal Spending')
plt.ylabel('Instructional Spending')
plt.legend(title='Elite Status')
plt.show()

# Boxplott för personliga utgifter mot elitstatus
plt.figure(figsize=(8, 6))
sns.boxplot(x="Elite", y="Personal", data=college, palette="coolwarm")
plt.title("Personal Spending vs Elite Status")
plt.xlabel("Elite Status")
plt.ylabel("Personal Spending")
plt.show()
```
